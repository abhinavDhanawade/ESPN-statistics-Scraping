# ESPN Statistic Scraping

```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
reposne1 = requests.get("https://stats.espncricinfo.com/ci/engine/records/team/series_results.html?class=2;id=6;type=team")
data1 = reposne1.text
cricket = BeautifulSoup(data1,'html.parser')
cric_tags= cricket.find_all('tbody')

matches={}
job_no = 0

for i in cric_tags:
    title= i.find_all('tr',class_="data1")
    for e in title:
        serie = e.find_all('td')
        o=[data.text for data in serie]
        Series_name=o[0]
        season=o[1]
        winner=o[3]
        margin=o[4]
        job_no+=1
        matches[job_no]=[Series_name,season,winner,margin]

matches_df= pd.DataFrame.from_dict(matches,orient="index",columns=['Series_name','season','winner','margin'])
matches_df
```

```python
matches_df.to_csv('B:/matches_df.csv')
```
